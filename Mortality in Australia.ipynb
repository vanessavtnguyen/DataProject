```python

def population_pyramid(refinedage_group):
    AgeClass = ['100','95-99','90-94','85-89','80-84','75-79','70-74','65-69','60-64','55-59','50-54', '45-49','40-44','35-39','30-34','25-29','20-24','15-19','10-14','5-9','0-4']
    bar_plot = sns.barplot(x='Males', y='Age', color='blue', data=refinedage_group, order=AgeClass)
    bar_plot = sns.barplot(x='Females', y='Age', color='red', data=refinedage_group, order=AgeClass)
    bar_plot.set(xlabel="Number of Deaths", ylabel="Age Group", title = "Number of Deaths by Sex and Age Group")
population_pyramid(refinedage_group)

```

## Leading Causes of Death 

```python
causes = retrievingdata('Leading_Causes_by_Age.csv')
```

```python
causes['Number of Deaths'] = causes['Number of Deaths'].astype(str).astype(int)
```

```python
all_ages = causes.loc[causes['Age'] == 'All ages']
all_ages
```

```python
my_range = range(1,len(all_ages.index)+1)
ordered_all_ages = all_ages.sort_values(by='Number of Deaths')
# Creates the Horizontal Lines 
plt.hlines(y=my_range, xmin=0, xmax=ordered_all_ages['Number of Deaths'], color="skyblue")
plt.plot(ordered_all_ages['Number of Deaths'], my_range, "o")
# Creates Categories on y-axis
plt.yticks(my_range, ordered_all_ages['Cause'])
# Labelling 
plt.ylabel("Cause of Death", fontsize=15)
plt.xlabel("Number of Deaths", fontsize=15)
plt.title("Top 20 Causes of Death in Australia 2018", loc='left', fontsize=24, fontweight="bold")
```

```python
heatmap_causes = causes[:-119][causes.Age != 'Under 1']
heatmap_causes
```

```python
fig, ax = plt.subplots()
data1 = pd.pivot_table(heatmap_causes, values='Per cent', index=['Cause'], columns='Age')
sns.heatmap(data1, cmap='PuBuGn')
plt.title("Percentage of Deaths by Main Causes for Ages 1 to 44 Australia 2018", fontsize=24, loc='left')
plt.xlabel("Age", fontsize=15)
plt.ylabel("Cause", fontsize=15)
plt.show()
```

From the heatmap the highest percentage of overall deaths of each age group by a certain cause are: 
- 1 to 14
    1. Land Transport Accidents
    2. Certain Conditions originating in the perintal period and deformations
    3. Brain Cancer 
- 15 to 24
    1. Suicide
    2. Land Transport Accidents
    3. Accidental Poisoning
- 25 to 44
    1. Suicide
    2. Accidental Poisoning
    3. Land Transport Accidents 

In all three age groups land transport accidents takes a high percentage of death despite it not being in the Top 20 Main Causes of death in 2018, however suicide takes first place in age groups 15 to 24 and 25 to 44.
